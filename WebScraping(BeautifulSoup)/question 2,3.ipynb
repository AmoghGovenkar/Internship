```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.imdb.com/')
page
```

```python
##display page source code
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
# scrap movie names
scraped_movies = soup.find_all('td', class_='titleColumn')
scraped_movies
```

```python
# parse movie names
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip(" ")
    movies.append(movie)
movies
```

```python
# scrap rating for movies
scraped_ratings = soup.find_all('td', class_='ratingColumn imdbRating')
scraped_ratings
```

```python
#parse ratings
ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n', '')
    ratings.apend(rating)
ratings
```

```python
data = pd.DataFrame()
data['Movie Names'] = movies
data['Ratings'] = ratings
data.head()
```
